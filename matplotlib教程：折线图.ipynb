## 导入包

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## 修复显示中文错误

```python
plt.rcParams['font.sans-serif']=['SimHei']   #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False#用来正常显示负号
```

## 生成数据

```python
x=np.linspace(-2,2,50)
y1=2*x+1 
y2=x**2
a=np.arange(1,20,3)
a1=np.random.randint(0,45,7)
a1
```

## 每个figure显示一张图片

```python
plt.figure(num='a',figsize=(20,1000))
```

## 绘图

```python
#每个plot显示一条线
plt.plot(x,y1,label='标签一')

#alpha代表的是透明度，小数
plt.plot(x, y2,label='标签二',color='red',linestyle=':',linewidth=10,alpha=1)  # 注意其中参数，记不住，看源码

#添加图例
plt.legend(loc='upper right') #loc参数可以为数值，1-10

#图片显示
plt.show()
```

## 为x,y轴以及标题添加标签

```python
plt.figure('第二个标签页')
plt.plot(a,a1)

#旋转角度
plt.xlabel('x标签')

plt.ylabel('y标签',fontsize='12',color='red')

plt.title('标题',fontsize='16')
#plt.savefig(r'./测试图片3.svg')   矢量图更适合放大

plt.show()
```

# 修改x，y轴的刻度，旋转角度

```python
plt.plot(a,a1)

#旋转角度
plt.xticks([2,4,6,8,10,12,14,19,20],rotation=45)

plt.yticks(range(min(a1),max(a1)+5,4))

plt.grid(alpha=0.8)

plt.show()
```
